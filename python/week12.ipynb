```python
# 실습문제 1번

import csv
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
max_temp = -999
min_temp = 999
temp_range = -999
max_date = ''
min_date = ''
temp_range_date = ''

for row in data:
    if row[-1] == '':
        row[-1] = -999
    row[-1] = float(row[-1])
    row[3] = float(row[3])
    
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]

    if min_temp > row[-1]:
        min_date = row[0]
        min_temp = row[-1]
        
    if (row[-1] - row[3]) > temp_range:
        temp_range = row[-1] - row[3]
        temp_range_date = row[0]
        
print('일교차가 가장 큰 날은', temp_range_date)
print('최저기온이 가장 낮은 날은', min_date)
print('최고기온이 가장 높은 날은', max_date)

f.close()
```

```python
# 실습문제 2번
# 2021년 12월

import csv
f = open('subway.csv')
data = csv.reader(f)

station_data = []
for row in data:
    station_data += data

data_len = len(station_data)
subway_line = [station_data[0][1]]
user = []
station_user = {}
for i in range(data_len-1):
    if station_data[i][1] != station_data[i+1][1]:
        subway_line.append(station_data[i+1][1])

for i in subway_line:
    for j in range(data_len):
        if station_data[j][1] == i:
            station_data[j][4] = station_data[j][4].replace(',','')
            station_data[j][5] = station_data[j][5].replace(',','')
            station_data[j][4] = int(station_data[j][4])
            station_data[j][5] = int(station_data[j][5])
            sum = station_data[j][4] + station_data[j][5]
            user.append(sum)
            station_user[station_data[j][3]] = sum 
    sorted_station_user = sorted(station_user.items(), key = lambda item:item[1])
    print(i)
    print('이용객이 가장 많은 역 순위')
    print('1.', sorted_station_user[-1][0])
    print('2.', sorted_station_user[-2][0])
    print('3.', sorted_station_user[-3][0])
    print('이용객이 가장 적은 역 순위')
    print('1.', sorted_station_user[0][0])
    print('2.', sorted_station_user[1][0])
    print('3.', sorted_station_user[2][0])
    user.clear()
    station_user.clear()
    sorted_station_user.clear()
    
f.close()
```
